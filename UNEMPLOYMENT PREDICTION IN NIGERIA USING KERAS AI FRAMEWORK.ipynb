```python
from keras import models
import numpy as np
from keras.layers import Dense, Dropout
from numpy import typename
from keras import activations
from keras import Sequential
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd
from keras.engine.input_layer  import Input
import keras
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import seaborn as sn
from keras.layers.advanced_activations import PReLU

import datetime as dt 
```

```python
#creating the keras model
#input_layer = keras.engine.input_layer.Input(shape = (32))

#model save counter
count = 0

datasets  =  pd.read_csv('dataset_new.csv')
#reshaping the network layer to have a shape of (1,None)
x_input_layer = datasets.Year.values.reshape(20,1)
y_output = datasets.drop(['Year'] ,  axis=1)
model  = Sequential()



# the input_shape = (1,) ==> Each set of the input to be passed into the network
#
model.add(Dense(40 ,kernel_initializer= 'uniform' ,  input_shape = (1,) , activation=PReLU(),bias_initializer= "zeros"))
model.add(Dense(40 , activation=PReLU(), kernel_initializer= "uniform" , bias_initializer= "zeros"))
#model.add(Dropout(0.2))
model.add(Dense(40 , activation=PReLU() , kernel_initializer= "uniform", bias_initializer= "zeros"))
model.add(Dense(40 , activation=PReLU(), kernel_initializer= "uniform",  bias_initializer= "zeros"))
#model.add(Dropout(0.2))
model.add(Dense(40 , activation=PReLU() , kernel_initializer= "uniform", bias_initializer= "zeros"))
model.add(Dense(40 , activation=PReLU() , kernel_initializer= "uniform", bias_initializer= "zeros"))
model.add(Dropout(0.2))
model.add(Dense(40 , activation=PReLU() , kernel_initializer= "uniform" ,bias_initializer= "zeros"))


#adding another dense layer of 64 input nuerons
model.add(Dense(units= 3 , activation="linear" , kernel_initializer="uniform"))

#compiling the model 
model.compile(loss=keras.losses.mse, metrics= ['accuracy'] , optimizer=keras.optimizers.Adam
             (lr=0.001, beta_1=0.9, beta_2=0.999 ,epsilon=1e-08, decay=0.0))

#split the data
x_train , x_test , y_train , y_test = train_test_split (x_input_layer,y_output,test_size = 0.30)

#scaling the feature
scalerX = MinMaxScaler()
x_train = scalerX.fit_transform(x_train)
x_test = scalerX.transform(x_test)

#scaling the feature
scalerY = MinMaxScaler()
y_train = scalerY.fit_transform(y_train)
y_test = scalerY.transform(y_test)


model.fit(x_train , y_train, batch_size= 2000 ,epochs= 10000,verbose=1)

#save the mode
try :
    save_model = model.to_json()
    with open("model.json{0}".format(dt.datetime.now()), "w") as json_model:
        json_model.write(save_model)
    
except Exception:
    print(Exception)
```
